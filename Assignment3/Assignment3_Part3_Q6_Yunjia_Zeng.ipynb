```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as py
import itertools  
import time
%matplotlib inline 
```

```python
ReLu = np.vectorize(lambda z: np.fmax(0,z))
sigmoid = lambda z: 1 / (1 + np.exp(-z))
```

## One layer One Node 

```python
V1=-1
Vb=1
w1=-1
b1=1
```

```python
x=np.linspace(-10,10,num=1000)
y=V1*ReLu(w1*x+b1)+Vb
py.figure(figsize=(16, 6))
py.plot(x, y, label='Training data',color='darkblue')
py.savefig('./relu_one_layer_one_node.jpg')
```

```python
x=np.linspace(-10,10,num=1000)
y=V1*sigmoid(w1*x+b1)+Vb
py.figure(figsize=(16, 6))
py.plot(x, y, label='Training data',color='darkblue')
py.savefig('./sigmoid_one_layer_one_node.jpg')
```

```python
Vb=0
V1=[-3,-1,1,3]
V2=[-5,-1,1,5]
w1=[-2,-1,1,2]
w2=[-2,-1,1,2]
b1=[10,20]
b2=[3,5]
```

```python
result = list(itertools.product(V1,V2,w1,w2,b1,b2))
```

## One layer Two Node 

```python
i=0
for V1, V2,w1,w2,b1,b2 in result:
    if i < 50:
        i+=1
        x=np.linspace(-10,10,num=1000)
        py.figure(figsize=(16, 6))
        y=V1*ReLu(w1*x+b1)+V2*ReLu(w2*x+b2)+Vb
        py.plot(x, y, label='Training data',color='darkblue')
        py.savefig('./relu12/one_layer_two_node'+str(i))
        py.close()
        py.figure(figsize=(16, 6))
        y=V1*sigmoid(w1*x+b1)+V2*sigmoid(w2*x+b2)+Vb
        py.plot(x, y, label='Training data',color='darkblue')
        py.savefig('./sigmoid12/one_layer_two_node'+str(i))
        py.close()
    else:
        break
```

## Two layers one Node 

```python
i=0
for V1, V2,w1,w2,b1,b2 in result:
    if i < 50:
        i+=1
        x=np.linspace(-10,10,num=1000)
        py.figure(figsize=(16, 6))
        y=V1*ReLu(w2*ReLu(w1*x+b1)+b2)+Vb
        py.plot(x, y, label='Training data',color='darkblue')
        py.savefig('./relu21/two_layer_one_node'+str(i))
        py.close()
        py.figure(figsize=(16, 6))
        y=V1*sigmoid(w2*sigmoid(w1*x+b1)+b2)+Vb
        py.plot(x, y, label='Training data',color='darkblue')
        py.savefig('./sigmoid21/two_layer_one_node'+str(i))
        py.close()
    else:
        break
```

## Two layers Two Node 

```python
Vb=0
V1=[-3,-1,1,3]
V2=[-4,-0.5,1,4]
w1=[-3,-1,1,3]
w2=[-3,-1,7,3]
w11=[-3,-4,1,3]
w22=[-2,-1,0.2,2]
w12=[-3,-0.1,1,3]
w21=[-2,-1,1,2]
b21=2
b22=30
b1=4
b2=5
result = list(itertools.product(V1,V2,w1,w2,w11,w22,w12,w21))
```

```python
i=0
for v1,v2,w1,w2,w11,w22,w12,w21 in result:
    if i < 50:
        i+=1
        x=np.linspace(-10,10,num=1000)
        py.figure(figsize=(16, 6))
        y=v1*ReLu(w11*ReLu(w1*x+b1)+w12*ReLu(w2*x+b2)+b21) +v2*ReLu(w21*ReLu(w1*x+b1)+w22*ReLu(w2*x+b2)+b22)+b2
        py.plot(x, y, label='Training data',color='darkblue')
        py.savefig('./relu22/two_layer_two_node'+str(i))
        py.close()
        py.figure(figsize=(16, 6))
        y=v1*sigmoid(w11*sigmoid(w1*x+b1)+w12*sigmoid(w2*x+b2)+b21) +v2*sigmoid(w21*sigmoid(w1*x+b1)+w22*sigmoid(w2*x+b2)+b22)+b2
        py.plot(x, y, label='Training data',color='darkblue')
        py.savefig('./sigmoid22/two_layer_two_node'+str(i))
        py.close()
    else:
        break
```
